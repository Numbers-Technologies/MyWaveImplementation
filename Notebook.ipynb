```python
%matplotlib inline

import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
audio_path: str = "test.wav"
x , sr = librosa.load(audio_path)
print(x.shape, sr)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
n0 = 9000
n1 = 9100
plt.figure(figsize=(14, 5))
plt.plot(x[n0:n1])
plt.grid()
```
